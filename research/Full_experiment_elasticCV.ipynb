```python
X = wine.drop(['quality'], axis=1)
Y = wine['quality']
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
x_train,x_test, y_train, y_test = train_test_split(X,Y,test_size=.2,random_state=42,shuffle=True)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
model = ElasticNet(alpha=1,l1_ratio=.5,random_state=42)
model.fit(X=x_train,y=y_train)
```

```python
predictions = model.predict(X=x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def eval(actual,predicted):
  print(mean_squared_error(actual,predicted))
  print(mean_absolute_error(actual,predicted))
  print(r2_score(actual,predicted))
```

```python
eval(y_test,predictions)
```
